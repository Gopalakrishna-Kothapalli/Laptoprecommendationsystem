```python
import pandas as pd
```

```python
persona_data = pd.read_csv('user_persona.csv')
```

```python
from data_transformer import CustomDataTransformer
```

```python
cdt = CustomDataTransformer(dataset=persona_data)
```

```python
modified_data = cdt.fit_transform()
```

```python
modified_data.head()
```

```python
X = modified_data.drop(' Suggested_Laptop_Category', axis=1)
Y = modified_data[' Suggested_Laptop_Category']
```

```python
new_X = X.drop(' Persona', axis= 1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
```

```python
lr.fit(new_X, Y)
dt.fit(new_X, Y)
rf.fit(new_X, Y)
```
